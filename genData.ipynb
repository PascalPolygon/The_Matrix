```python
# print(data.shape)
# print(data[1][:,0])
training_data = np.empty([64, 15000])
testing_data = np.empty([64, 15000])
for i in range(data.shape[0]):
    if data[i][0][0] == 3 or data[i][0][0] == 7 or data[i][0][0] == 4 or data[i][0][0] == 8 or data[i][0][0] == 5 or data[i][0][0] == 9 or data[i][0][0] == 6 or data[i][0][0] == 10:
        # copy training data
        training_data = np.append(training_data, data[i], axis=0)
    elif data[i][0][0] == 0:
        # This would be the first matrix, we don't need it
        continue
    else:
        # copy testing data
        testing_data = np.append(testing_data, data[i], axis=0)
print(training_data.shape)
print(testing_data.shape)
```
